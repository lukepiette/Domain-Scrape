```python
import requests
import time
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
```

```python
def getDomains(ext):
    txt = open('1syllableverbs.txt','r').read()
    verbList = [i for i in txt.split('\n')]
    agent = {"User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/83.0.4103.61 Chrome/83.0.4103.61 Safari/537.36'}
    verbDict = {}
    
    for verb in verbList:
        try:
            URL = f'http://{verb}{ext}.com'
            response = requests.get(URL,headers=agent)
            verbDict[verb] = 'taken'
            print(f'{verb}{ext}.com TAKEN')
        except:
            verbDict[verb] = 'available'
            print(f'\n{verb}{ext}.com AVAILABLE!\n')
    
    return verbDict
```

```python
def requestDomain(verb,ext):
    agent = {"User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/83.0.4103.61 Chrome/83.0.4103.61 Safari/537.36'}
    
    try:
        URL = f'http://{verb}{ext}.com'
        response = requests.get(URL,headers=agent)
        return 'taken'
    except:
        verbDict[verb] = 'available'
        return 'available'
```

```python
def headlessGodaddyGet(URList):
    driver = webdriver.Chrome(ChromeDriverManager().install(),options=chrome_options)
    
    for URL in URList:
        driver.get(URL)
        HTML = str(BeautifulSoup(driver.page_source.encode("utf-8")))
        print(HTML)
        driver.quit()
```
